# qikit.visualization.plot_state_qsphere

**plot_state_sphere** is a function used to give the qshere representation of the quantum state.
- *Size of the points* represents the *probability* of the corresponding term in the state
- *Color* represents *phase* of the state

**qiskit.visualization.plot_state_qsphere(state, figsize=None, ax=None, show_state_labels=True, show_state_phases=False, use_degrees=False, \*, filename=None)**



The function parameters: <br>
| Parameters | Description | typedef |
| :---: | :---: | :---: |
| state | N qubit quantum state | StateVector/DensityMatrix/ndarray|
| figsize | Figure size (inches) | tuple |
| ax | Axes object (None => Matplotlib) | matplotlib.axes.Axes |
| show_state_labels | Whether to show labels for each basis state | bool |
| show_state_phases | Whether to show the phase for each basis state | bool |
| use_degrees | Degrees if not Radians | bool |

- The function returns "matplotlib figure instance" if the "ax" kwarg is not set
- Return type => matplotlib.figure.Figure

The function might raise 3 error: <br>
| Error | Description |
| :---: | :---: |
| MissingOptionalLibraryError | Requires Matplotlib |
| VisualizationError | Wrong Input, invalid N-qubit system |
| QiskitError | Wrong Input Statevector => not valid dimensions |


```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_state_qsphere

qc = QuantumCircuit(2)
qc.h(0)
qc.cx(0, 1)
# A bell state created

state = Statevector(qc)
plot_state_qsphere(state)
```

```python
from qiskit.quantum_info import DensityMatrix
import numpy as np
from qiskit import QuantumCircuit
from qiskit.visualization import plot_state_qsphere

qc = QuantumCircuit(2)
qc.h([0, 1])
qc.cz(0, 1)
qc.ry(np.pi/3, 0)
qc.rx(np.pi/5, 1)
qc.z(1)

matrix = DensityMatrix(qc)
plot_state_qsphere(matrix, show_state_labels=True, show_state_phases=True, use_degrees=True)
```
